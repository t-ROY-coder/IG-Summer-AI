```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_meal = pd.read_csv('meal_info.csv') # I have it in the same directory as the notebook
df_meal.head()
```

```python
df_center = pd.read_csv('fulfilment_center_info.csv')
df_center.head()
```

```python
df_food = pd.read_csv('train.csv')
df_food.head()
```

```python
df = pd.merge_ordered(df_food, df_meal, on = 'meal_id')
df = pd.merge_ordered(df, df_center, on = 'center_id')
df.head()
```

```python
table = pd.pivot_table(data=df,index='category',values='num_orders',aggfunc=np.sum)
table
# This creates spreadsheet-style pivot table containing the sum of 'num_orders' for each type of 'category'
```

```python
orders = []
for i in range(table.size):
    orders.append(table.num_orders[i])
plt.bar(table.index, orders)
plt.xlabel('Food Items')
plt.ylabel('Quantity Sold')
plt.title('Most popular food')
plt.xticks(rotation=70) # This rotates the x-labels by 70 deg from left, to fit in the graph and it readable
plt.show()
```

```python
df['revenue'] = df.checkout_price * df.num_orders
df['month'] = ((df.week-1)//4)+1
df.head()
```

```python
table2 = pd.pivot_table(data=df,index='month',values='revenue',aggfunc=np.sum)
month = [ele for ele in table2.index]
month_order = [ele for ele in table2.revenue]
```

```python
table3 = pd.pivot_table(data=df,index='week',values='revenue',aggfunc=np.sum)
week = [ele for ele in table3.index]
week_order = [ele for ele in table3.revenue]
```

```python
fig, ax = plt.subplots(1,2,figsize = (20,5))

ax[0].plot(week, week_order)
ax[1].plot(month, month_order)
ax[0].set_title('Weekly Income')
ax[1].set_title('Monthly Income')
ax[0].set_xlabel('Week')
ax[0].set_ylabel('Revenue')
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Revenue')

plt.show()
```
